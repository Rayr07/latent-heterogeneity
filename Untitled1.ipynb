## Prepare data for time windowing

### Subtask:
Ensure the data is sorted by a time-based column if one exists, or assume the data is a stream and apply a simple rolling window based on row index.


**Reasoning**:
Check for a time-based column in the DataFrame and sort if found, or proceed with the current index order if not.



```python
# Check for a time-based column. Common names include 'timestamp', 'time', 'datetime', etc.
# Based on the available columns in df (id, dur, proto, service, state, spkts, dpkts, sbytes, dbytes, rate, ...),
# 'dur' (duration) might represent the duration of a flow, which is related to timing.
# However, there is no explicit timestamp column to sort by event time.
# Assuming the data is ordered by the flow start time or sequence, the current index order is acceptable for rolling window operations.
# If a timestamp column were present, the code would look like this:
# if 'timestamp_column_name' in df.columns:
#     df_sorted = df.sort_values(by='timestamp_column_name').copy()
# else:
#     df_sorted = df.copy() # Data is assumed to be in stream order

# Since no explicit timestamp column is available, we assume the data is already in a stream order based on index.
# No sorting is needed for this step. The DataFrame 'df' is already in a suitable order.
print("No explicit timestamp column found. Assuming data is in stream order based on index.")
```

## Define rolling window

### Subtask:
Determine the window size ($T$) and how the window will slide (e.g., overlapping or non-overlapping).


**Reasoning**:
Define the window size for the rolling calculations.



```python
# Define the window size for the rolling calculations.
# A window size of 100 means that for each data point,
# the rolling statistics will be calculated using the current data point and the previous 99 data points.
window_size = 100

print(f"Rolling window size (T) defined as: {window_size}")

# For this task, we are using a simple rolling window based on row index.
# This means the window slides one step at a time, creating overlapping windows.
print("Using a simple rolling window based on row index (overlapping windows).")
```

## Calculate high-order statistics (hos)

### Subtask:
For each rolling window and each selected feature, calculate the Mean, Variance (or Standard Deviation), Skewness, and Kurtosis.


**Reasoning**:
Calculate the rolling mean, variance, skewness, and kurtosis for each selected feature using the defined window size and store them in new columns.



```python
# Calculate rolling statistics for each selected feature
for feature in timing_rhythm_volume_features:
    df_selected_features[f'{feature}_rolling_mean'] = df_selected_features[feature].rolling(window=window_size).mean()
    df_selected_features[f'{feature}_rolling_variance'] = df_selected_features[feature].rolling(window=window_size).var()
    df_selected_features[f'{feature}_rolling_skew'] = df_selected_features[feature].rolling(window=window_size).skew()
    df_selected_features[f'{feature}_rolling_kurtosis'] = df_selected_features[feature].rolling(window=window_size).kurt()

# Display the first few rows with the new rolling features
display(df_selected_features.head())
```

## Calculate rolling autocorrelation

### Subtask:
For each rolling window and each selected feature, calculate the rolling autocorrelation at a specified lag ($N$).


**Reasoning**:
The previous command failed because the `rolling` object in pandas does not have a direct `autocorr` method. To calculate rolling autocorrelation, we need to use `rolling().corr()` between the feature and its lagged version.



```python
# Define the lag size for rolling autocorrelation
lag_size = 5

# Calculate rolling autocorrelation for each selected feature
for feature in timing_rhythm_volume_features:
    # Create a lagged version of the feature
    lagged_feature = df_selected_features[feature].shift(lag_size)
    # Calculate rolling correlation between the feature and its lagged version
    df_selected_features[f'{feature}_rolling_autocorr_lag{lag_size}'] = df_selected_features[feature].rolling(window=window_size).corr(lagged_feature)

# Display the first few rows with the new rolling autocorrelation features
display(df_selected_features.head())
```

## Get anomaly scores

### Subtask:
Use the trained Isolation Forest model to get the anomaly scores for each data point in `df_engineered_features` using the `decision_function`.


**Reasoning**:
Calculate the anomaly scores for each data point in the engineered features DataFrame using the trained Isolation Forest model's decision_function.



```python
# Use the trained Isolation Forest model to get the anomaly scores
# The decision_function returns the anomaly score for each sample.
# Lower values indicate more anomalous samples.
anomaly_scores = model.decision_function(df_engineered_features)

print("Anomaly scores calculated.")
```

## Determine threshold

### Subtask:
Set the contamination parameter and calculate the decision threshold based on the anomaly scores and the contamination factor.


**Reasoning**:
Define the contamination parameter and calculate the decision threshold using the quantile method on the anomaly scores.



```python
# Define the contamination parameter (proportion of outliers in the data set)
# We can set this based on prior knowledge or an assumption about the data.
# For this example, let's assume 1% of the data is anomalous.
contamination = 0.01

# Calculate the decision threshold.
# The decision_function returns lower values for anomalies.
# We need to find the value below which 'contamination' proportion of the data falls.
decision_threshold = pd.Series(anomaly_scores).quantile(contamination)

print(f"Contamination parameter set to: {contamination}")
print(f"Calculated decision threshold: {decision_threshold}")
```

## Predict anomalies

### Subtask:
Use the determined threshold to predict anomalies (outliers) based on the anomaly scores.


**Reasoning**:
Compare the anomaly scores with the determined decision threshold to classify each data point as normal or anomalous and store these labels in a new variable.



```python
# Predict anomalies based on the decision threshold
# Values <= decision_threshold are anomalies (-1), values > decision_threshold are normal (1)
predicted_labels = [1 if score > decision_threshold else -1 for score in anomaly_scores]

print("Anomaly prediction complete.")
```

## Add predictions to dataframe

Add the anomaly scores and predicted labels to a new DataFrame or the existing `df_engineered_features` DataFrame for further analysis and evaluation.

Add the anomaly scores and predicted labels as new columns to the df_engineered_features DataFrame and display the head to verify.

## Extract true labels

Get the true anomaly labels from the original DataFrame (`df`).

### Subtask:
Calculate the Macro F1 Score using the true labels and the predicted anomaly labels.

```python
from sklearn.metrics import f1_score

# Calculate the Macro F1 Score
macro_f1 = f1_score(true_labels_aligned, df_engineered_features['predicted_anomaly_label'], average='macro')

# Print the Macro F1 Score
print(f"Macro F1 Score: {macro_f1}")
```

## Calculate auc-pr

### Subtask:
Calculate the Area Under the Precision-Recall Curve (AUC-PR) using the true labels and the anomaly scores.


**Reasoning**:
Import the necessary functions for calculating AUC-PR and then calculate the precision, recall, and AUC-PR using the true labels and anomaly scores.



```python
from sklearn.metrics import precision_recall_curve, auc

# Calculate precision and recall
# precision_recall_curve expects the positive class to be 1.
# Our true_labels_aligned has -1 for anomaly (positive class in our context) and 1 for normal.
# We need to convert the true labels for this function: -1 -> 1 (positive), 1 -> 0 (negative)
true_labels_for_pr_curve = true_labels_aligned.apply(lambda x: 1 if x == -1 else 0)

precision, recall, thresholds = precision_recall_curve(true_labels_for_pr_curve, -df_engineered_features['anomaly_score'])

# Calculate AUC-PR
auc_pr = auc(recall, precision)

print(f"Area Under the Precision-Recall Curve (AUC-PR): {auc_pr}")
```

## Report metrics

### Subtask:
Display the calculated Macro F1 Score and AUC-PR.


**Reasoning**:
Display the calculated Macro F1 Score and AUC-PR from the previously computed variables.



```python
print(f"Macro F1 Score: {macro_f1}")
print(f"Area Under the Precision-Recall Curve (AUC-PR): {auc_pr}")
```

**Reasoning**:
Based on the distribution of the 'dur' column, which is heavily skewed towards smaller values, quantile-based binning is a suitable approach to create intervals with roughly equal numbers of data points. This will help ensure that each interval is represented in the time series graph.



```python
# Determine the number of intervals
num_intervals = 10 # Choose a reasonable number of intervals

# Create quantile-based bins for the 'dur' column
# This method ensures each bin has approximately the same number of data points
df['dur_interval'], bins = pd.qcut(df['dur'], q=num_intervals, labels=False, retbins=True, duplicates='drop')

# Create labels for the intervals based on the bin edges
# Format the bin edges for better readability
bin_labels = [f'{bins[i]:.4f} - {bins[i+1]:.4f}' for i in range(len(bins)-1)]

# Map the numerical labels to the descriptive bin labels
# If duplicates='drop' resulted in fewer bins than num_intervals, adjust bin_labels
if len(bin_labels) != df['dur_interval'].nunique():
    # Re-calculate bin_labels based on the actual unique intervals created by qcut
    unique_intervals = sorted(df['dur_interval'].dropna().unique())
    actual_bins = [bins[i] for i in sorted(np.unique(np.digitize(df['dur'], bins) - 1))]
    # Add the max value to the actual_bins if it's not the upper bound of the last bin
    if actual_bins[-1] < df['dur'].max():
        actual_bins.append(df['dur'].max())

    bin_labels = [f'{actual_bins[i]:.4f} - {actual_bins[i+1]:.4f}' for i in range(len(actual_bins)-1)]
    # Handle the edge case where qcut might assign the max value to the last bin index
    if len(bin_labels) < len(unique_intervals):
         bin_labels.append(f'> {actual_bins[-1]:.4f}') # or handle as appropriate


# Add the descriptive labels to the DataFrame
# Ensure the mapping is correct if some quantile bins were dropped due to duplicates
interval_mapping = {i: label for i, label in enumerate(bin_labels)}
df['dur_interval_label'] = df['dur_interval'].map(interval_mapping)


print(f"Created {df['dur_interval'].nunique()} duration intervals using quantile-based binning.")
print("\nFirst few rows with the new 'dur_interval' and 'dur_interval_label' columns:")
display(df[['dur', 'dur_interval', 'dur_interval_label']].head())

print("\nValue counts for each duration interval:")
print(df['dur_interval_label'].value_counts().sort_index())
```
