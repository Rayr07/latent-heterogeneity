```python
import pandas as pd

manifest_path = "/content/drive/MyDrive/permanent_data_folder/gdc_manifest.2026-01-14.172833.txt"

df = pd.read_csv(manifest_path, sep="\t")
total_bytes = df["size"].sum()

print(f"Total required: {total_bytes / (1024**3):.2f} GB")
```

```python
import os

drive_dir = "/content/drive/MyDrive/permanent_data_folder/TCGA_BRCA_data"

downloaded_bytes = 0
for root, _, files in os.walk(drive_dir):
    for f in files:
        downloaded_bytes += os.path.getsize(os.path.join(root, f))

print(f"Downloaded so far: {downloaded_bytes / (1024**3):.2f} GB")

percent = 100 * downloaded_bytes / total_bytes
print(f"Progress: {percent:.2f}%")
```
