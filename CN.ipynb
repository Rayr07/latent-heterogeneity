## Implement dumb queue simulation

### Subtask:
Build a simulation using Python's `queue.Queue` (FIFO) to represent the baseline queue.


**Reasoning**:
Implement the FIFO queue simulation function as described in the instructions.



```python
import queue
import time

def simulate_fifo_queue(tasks, processing_rate):
    """
    Simulates a FIFO queue for task processing.

    Args:
        tasks: A list of tuples, where each tuple is (arrival_time, base_priority).
               Note: base_priority is not used in FIFO but kept for consistency.
        processing_rate: The number of tasks that can be processed per unit of time.

    Returns:
        A dictionary containing simulation metrics:
        - 'total_tasks_processed': The total number of tasks processed.
        - 'average_waiting_time': The average time tasks spent waiting in the queue.
    """
    q = queue.Queue()
    current_time = 0
    total_waiting_time = 0
    tasks_processed_count = 0

    # Sort tasks by arrival time
    sorted_tasks = sorted(tasks, key=lambda x: x[0])

    task_index = 0
    while tasks_processed_count < len(sorted_tasks) or not q.empty():
        # Add arriving tasks to the queue
        while task_index < len(sorted_tasks) and sorted_tasks[task_index][0] <= current_time:
            q.put(sorted_tasks[task_index])
            task_index += 1

        # Process tasks from the queue
        if not q.empty():
            # Simulate processing based on the rate
            # In a simple simulation, we can process 'processing_rate' tasks instantly
            # or advance time based on the processing rate.
            # Let's assume processing_rate is tasks per unit time, and we process one task per time unit for simplicity
            # A more complex simulation would involve fractional time steps or event-based simulation.
            # For this simple FIFO, let's process one task if available and advance time.
            task_start_processing_time = current_time
            task = q.get()
            waiting_time = task_start_processing_time - task[0]
            total_waiting_time += waiting_time
            tasks_processed_count += 1
            current_time += 1/processing_rate # Advance time by the processing time of one task
        else:
            # If the queue is empty, advance time to the next task arrival or simply by a small step
            if task_index < len(sorted_tasks):
                current_time = sorted_tasks[task_index][0]
            else:
                 # If no more tasks are arriving and queue is empty, simulation ends
                 break


    average_waiting_time = total_waiting_time / tasks_processed_count if tasks_processed_count > 0 else 0

    return {
        'total_tasks_processed': tasks_processed_count,
        'average_waiting_time': average_waiting_time,
    }

print("FIFO queue simulation function defined.")
```

## Implement metabolic queue simulation

### Subtask:
Build a simulation using Python's `heapq` (priority queue) that utilizes the `dynamic_priority` calculated in Step 1.


**Reasoning**:
Implement the simulation function for the metabolic queue using heapq, incorporating the dynamic priority calculation and handling task processing and time advancement.



```python
import heapq

def calculate_dynamic_priority(base_priority, time_in_queue):
    """
    Calculates the dynamic priority of a task.

    Args:
        base_priority: The base priority of the task.
        time_in_queue: The time the task has spent in the queue.

    Returns:
        The calculated dynamic priority.
    """
    # Lower dynamic priority is higher priority in heapq
    dynamic_priority = base_priority - time_in_queue
    return dynamic_priority

def simulate_metabolic_queue(tasks, processing_rate):
    """
    Simulates a Metabolic queue using a priority queue (heapq).

    Args:
        tasks: A list of tuples, where each tuple is (arrival_time, base_priority).
        processing_rate: The number of tasks that can be processed per unit of time.

    Returns:
        A dictionary containing simulation metrics:
        - 'total_tasks_processed': The total number of tasks processed.
        - 'average_waiting_time': The average time tasks spent waiting in the queue.
    """
    # Priority queue stores tuples: (dynamic_priority, arrival_time, base_priority, task_id)
    # heapq is a min-heap, so lower dynamic_priority means higher actual priority.
    pq = []
    current_time = 0
    total_waiting_time = 0
    tasks_processed_count = 0
    task_id_counter = 0

    # Sort tasks by arrival time
    sorted_tasks = sorted(tasks, key=lambda x: x[0])
    task_index = 0

    while tasks_processed_count < len(sorted_tasks) or pq:
        # Add arriving tasks to the priority queue
        while task_index < len(sorted_tasks) and sorted_tasks[task_index][0] <= current_time:
            arrival_time, base_priority = sorted_tasks[task_index]
            initial_dynamic_priority = calculate_dynamic_priority(base_priority, 0) # Time in queue is 0 at arrival
            # Store (dynamic_priority, arrival_time, base_priority, task_id)
            heapq.heappush(pq, (initial_dynamic_priority, arrival_time, base_priority, task_id_counter))
            task_id_counter += 1
            task_index += 1

        # Process tasks from the priority queue
        if pq:
            # When popping, recalculate dynamic priority based on current time
            # This handles tasks that have been waiting and whose priority has increased
            # We need to peek at the top without removing it to check its current priority
            # A more efficient way is to recalculate priority just before processing
            # Let's pop and then calculate the *actual* dynamic priority at current_time

            # Get the task that was *initially* highest priority
            # (priority, arrival_time, base_priority, task_id) = heapq.heappop(pq)

            # To use the most up-to-date priority, we should calculate it at the moment of processing
            # The tuple in the heap is (initial_dynamic_priority, arrival_time, base_priority, task_id)
            # Let's process one task if available and advance time.
            task_to_process_data = heapq.heappop(pq)
            # The stored priority in the heap might be outdated, calculate the current dynamic priority
            # using the arrival_time and current_time
            stored_arrival_time = task_to_process_data[1]
            stored_base_priority = task_to_process_data[2]
            # stored_task_id = task_to_process_data[3] # Not strictly needed for metrics

            # Calculate the *actual* waiting time before processing starts
            waiting_time = current_time - stored_arrival_time
            total_waiting_time += waiting_time
            tasks_processed_count += 1

            # Advance time by the processing time of one task
            current_time += 1/processing_rate

            # Note: In this simple simulation, we don't re-add tasks with updated priorities.
            # The re-calculation happens implicitly because the task with the lowest
            # (base_priority - time_in_queue) at the *moment of popping* is processed.
            # The initial dynamic priority stored in the heap is primarily for the heap's
            # internal ordering at the time of insertion, but we use the real-time
            # waiting time for calculating total waiting time.
            # A more complex simulation might periodically update priorities in the heap,
            # but that's computationally expensive with heapq. The current approach
            # effectively processes the task that has the highest priority *now*.

        else:
            # If the queue is empty, advance time to the next task arrival
            if task_index < len(sorted_tasks):
                current_time = max(current_time, sorted_tasks[task_index][0])
            else:
                 # If no more tasks are arriving and queue is empty, simulation ends
                 break


    average_waiting_time = total_waiting_time / tasks_processed_count if tasks_processed_count > 0 else 0

    return {
        'total_tasks_processed': tasks_processed_count,
        'average_waiting_time': average_waiting_time,
    }

print("Metabolic queue simulation function defined.")
```

## Modify simulation functions

Update both the FIFO and Metabolic queue simulation functions to handle the larger number of tasks, record queue length over time, and record individual video packet waiting times.

```python
import queue
import time
import heapq
import numpy as np # Need numpy for isnan check if any NaNs somehow creep in (though cleaning should prevent this)


def calculate_dynamic_priority(base_priority, time_in_queue, metabolic_weight):
    """
    Calculates the dynamic priority of a task using a metabolic weight.

    Args:
        base_priority: The base priority of the task.
        time_in_queue: The time the task has spent in the queue.
        metabolic_weight: The weight applied to time in queue for dynamic priority.

    Returns:
        The calculated dynamic priority.
    """
    # Lower dynamic priority is higher priority in heapq
    # Ensure time_in_queue is not negative
    effective_time_in_queue = max(0, time_in_queue)
    dynamic_priority = base_priority - (effective_time_in_queue * metabolic_weight)
    return dynamic_priority


def simulate_fifo_queue(tasks, processing_rate):
    """
    Simulates a FIFO queue for task processing, recording queue length and video waiting times.

    Args:
        tasks: A list of tuples, where each tuple is (arrival_time, base_priority, task_type).
               Note: base_priority and task_type are not used for processing order in FIFO.
        processing_rate: The number of tasks that can be processed per unit of time.

    Returns:
        A dictionary containing simulation metrics and recorded data:
        - 'total_tasks_processed': The total number of tasks processed.
        - 'average_waiting_time': The average time tasks spent waiting in the queue.
        - 'fifo_queue_lengths': List of queue lengths recorded over time.
        - 'fifo_video_waiting_times': List of waiting times for 'Video' tasks.
        - 'simulation_times': List of time points at which metrics were recorded.
    """
    q = queue.Queue()
    current_time = 0
    total_waiting_time = 0
    tasks_processed_count = 0

    fifo_queue_lengths = []
    fifo_video_waiting_times = []
    simulation_times = []

    # Sort tasks by arrival time
    sorted_tasks = sorted(tasks, key=lambda x: x[0])

    task_index = 0
    # Simulate until all tasks have arrived and the queue is empty
    while tasks_processed_count < len(sorted_tasks) or not q.empty():
        # Record queue length at the current time step
        fifo_queue_lengths.append(q.qsize())
        simulation_times.append(current_time)

        # Add arriving tasks to the queue
        while task_index < len(sorted_tasks) and sorted_tasks[task_index][0] <= current_time:
            # Task format is (arrival_time, base_priority, task_type)
            q.put(sorted_tasks[task_index])
            task_index += 1

        # Process tasks from the queue
        if not q.empty():
            # Simulate processing based on the rate
            task_arrival_time, _, task_type = q.get() # Get task details
            waiting_time = current_time - task_arrival_time
            total_waiting_time += waiting_time
            tasks_processed_count += 1

            if task_type == 'Video':
                fifo_video_waiting_times.append(waiting_time)

            # Advance time by the processing time of one task
            # Assuming processing time is 1/processing_rate for each task
            current_time += 1.0/processing_rate
        else:
            # If the queue is empty, advance time to the next task arrival
            if task_index < len(sorted_tasks):
                current_time = sorted_tasks[task_index][0]
            else:
                 # If no more tasks are arriving and queue is empty, simulation ends
                 break

    # Append final queue length (should be 0)
    fifo_queue_lengths.append(q.qsize())
    simulation_times.append(current_time)


    average_waiting_time = total_waiting_time / tasks_processed_count if tasks_processed_count > 0 else 0

    return {
        'total_tasks_processed': tasks_processed_count,
        'average_waiting_time': average_waiting_time,
        'fifo_queue_lengths': fifo_queue_lengths,
        'fifo_video_waiting_times': fifo_video_waiting_times,
        'simulation_times': simulation_times
    }


def simulate_metabolic_queue(tasks, processing_rate, metabolic_weight):
    """
    Simulates a Metabolic queue using a priority queue (heapq), recording queue length and video waiting times.

    Args:
        tasks: A list of tuples, where each tuple is (arrival_time, base_priority, task_type).
        processing_rate: The number of tasks that can be processed per unit of time.
        metabolic_weight: The weight used in the dynamic priority calculation.

    Returns:
        A dictionary containing simulation metrics and recorded data:
        - 'total_tasks_processed': The total number of tasks processed.
        - 'average_waiting_time': The average time tasks spent waiting in the queue.
        - 'metabolic_queue_lengths': List of queue lengths recorded over time.
        - 'metabolic_video_waiting_times': List of waiting times for 'Video' tasks.
        - 'simulation_times': List of time points at which metrics were recorded.
    """
    # Priority queue stores tuples: (dynamic_priority, arrival_time, base_priority, task_type, task_id)
    # heapq is a min-heap, so lower dynamic_priority means higher actual priority.
    pq = []
    current_time = 0
    total_waiting_time = 0
    tasks_processed_count = 0
    task_id_counter = 0 # To uniquely identify tasks

    metabolic_queue_lengths = []
    metabolic_video_waiting_times = []
    simulation_times = [] # Use a single list for time points for both simulations

    # Sort tasks by arrival time
    sorted_tasks = sorted(tasks, key=lambda x: x[0])
    task_index = 0

    # Simulate until all tasks have arrived and the queue is empty
    while tasks_processed_count < len(sorted_tasks) or pq:
        # Record queue length at the current time step
        metabolic_queue_lengths.append(len(pq))
        simulation_times.append(current_time) # Append time here

        # Add arriving tasks to the priority queue
        while task_index < len(sorted_tasks) and sorted_tasks[task_index][0] <= current_time:
            arrival_time, base_priority, task_type = sorted_tasks[task_index]
            # Calculate initial dynamic priority (time in queue is 0 at arrival)
            initial_dynamic_priority = calculate_dynamic_priority(base_priority, 0, metabolic_weight)
            # Store (dynamic_priority, arrival_time, base_priority, task_type, task_id)
            heapq.heappush(pq, (initial_dynamic_priority, arrival_time, base_priority, task_type, task_id_counter))
            task_id_counter += 1
            task_index += 1

        # Process tasks from the priority queue
        if pq:
            # When processing, we need the task that has the highest dynamic priority *now*.
            # The priority stored in the heap is based on the time the task was inserted.
            # A correct implementation requires re-evaluating priorities or using a different
            # priority queue structure that supports efficient priority updates.
            # However, heapq doesn't efficiently support priority updates.
            # A common workaround is to insert updated priority items and ignore older ones.
            # For this simulation, let's approximate by recalculating priority of the TOP element
            # before processing. This is not perfectly accurate if a task's priority increases
            # while it's *not* at the top, but it's a reasonable simplification for demonstration.
            # A more accurate approach would involve a custom priority queue or re-heapifying periodically.

            # Let's re-evaluate the dynamic priority of all tasks in the queue at current_time
            # This is computationally more expensive but necessary for correct metabolic behavior.
            # Create a temporary list to hold tasks with updated priorities
            temp_pq = []
            while pq:
                # (priority, arrival_time, base_priority, task_type, task_id) = heapq.heappop(pq) # This is the old priority
                # We need the original arrival_time and base_priority to calculate the current dynamic priority
                # Let's rebuild the heap with current priorities
                # To avoid modifying the heap while iterating, pop all, update, then push back.
                (old_priority, arrival_time, base_priority, task_type, task_id) = heapq.heappop(pq)
                current_dynamic_priority = calculate_dynamic_priority(base_priority, current_time - arrival_time, metabolic_weight)
                heapq.heappush(temp_pq, (current_dynamic_priority, arrival_time, base_priority, task_type, task_id))

            # Replace the old heap with the new one
            pq = temp_pq

            # Now, the task with the highest current dynamic priority is at the top
            # Get the task with the highest priority (lowest dynamic_priority value)
            (dynamic_priority, arrival_time, base_priority, task_type, task_id) = heapq.heappop(pq)

            # Calculate the actual waiting time before processing starts
            waiting_time = current_time - arrival_time
            total_waiting_time += waiting_time
            tasks_processed_count += 1

            if task_type == 'Video':
                metabolic_video_waiting_times.append(waiting_time)

            # Advance time by the processing time of one task
            current_time += 1.0/processing_rate
        else:
            # If the queue is empty, advance time to the next task arrival
            if task_index < len(sorted_tasks):
                current_time = max(current_time, sorted_tasks[task_index][0])
            else:
                 # If no more tasks are arriving and queue is empty, simulation ends
                 break

    # Append final queue length (should be 0)
    metabolic_queue_lengths.append(len(pq))
    simulation_times.append(current_time)


    average_waiting_time = total_waiting_time / tasks_processed_count if tasks_processed_count > 0 else 0

    return {
        'total_tasks_processed': tasks_processed_count,
        'average_waiting_time': average_waiting_time,
        'metabolic_queue_lengths': metabolic_queue_lengths,
        'metabolic_video_waiting_times': metabolic_video_waiting_times,
        'simulation_times': simulation_times # Return simulation times from one simulation (they should be similar)
    }

print("Updated simulation functions to record queue length and video waiting times.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import deque
import time
# We don't need 'copy' anymore, it failed us.

# --- 0. DATA CLEANING (Run once) ---
print("Loading and cleaning data...")
try:
    df = pd.read_csv('/content/consolidated_traffic_data.csv')
except FileNotFoundError:
    print("Error: CSV not found. Please re-run the Kaggle download cells.")
    raise

def simplify_label(row):
    label = row['traffic_type']
    if 'STREAMING' in label: return 'Video'
    elif 'CHAT' in label: return 'Chat'
    elif 'VOIP' in label: return 'VoIP'
    elif 'FT' in label: return 'File_Transfer'
    elif 'P2P' in label: return 'File_Transfer'
    elif 'MAIL' in label: return 'Email'
    elif 'BROWSING' in label: return 'Browsing'
    else: return 'Other'
df['Label'] = df.apply(simplify_label, axis=1)
print(df['Label'].value_counts().head())

# --- SIMULATION PARAMETERS ---
TASK_COUNT = 2000
PROCESSING_RATE = 1.0
PROCESS_TIME_PER_TASK = 1.0 / PROCESSING_RATE
ARRIVAL_LAMBDA = 0.95
BASE_PRIORITY = {
    'Video': 100, 'VoIP': 90, 'Chat': 80,
    'Browsing': 30, 'File_Transfer': 10, 'Email': 5, 'Other': 1
}
METABOLIC_WEIGHT = 0.01

# --- SIMULATION FUNCTIONS (unchanged) ---
def simulate_fifo_queue(task_list, proc_time):
    current_time = 0.0
    processor_free_time = 0.0
    task_index = 0
    waiting_queue = deque()
    completed_tasks = []
    while task_index < len(task_list) or waiting_queue:
        next_arrival_time = task_list[task_index]['arrival_time'] if task_index < len(task_list) else float('inf')
        if not waiting_queue: current_time = next_arrival_time
        else: current_time = processor_free_time
        if current_time == float('inf'): break
        while task_index < len(task_list) and task_list[task_index]['arrival_time'] <= current_time:
            waiting_queue.append(task_list[task_index])
            task_index += 1
        if waiting_queue:
            task_to_process = waiting_queue.popleft()
            task_to_process['wait_time'] = current_time - task_to_process['arrival_time']
            completed_tasks.append(task_to_process)
            processor_free_time = current_time + proc_time
    return {'completed_tasks': completed_tasks}

def simulate_metabolic_queue(task_list, base_priorities, weight, proc_time):
    current_time = 0.0
    processor_free_time = 0.0
    task_index = 0
    waiting_queue = []
    completed_tasks = []
    while task_index < len(task_list) or waiting_queue:
        next_arrival_time = task_list[task_index]['arrival_time'] if task_index < len(task_list) else float('inf')
        if not waiting_queue: current_time = next_arrival_time
        else: current_time = processor_free_time
        if current_time == float('inf'): break
        while task_index < len(task_list) and task_list[task_index]['arrival_time'] <= current_time:
            waiting_queue.append(task_list[task_index])
            task_index += 1
        if waiting_queue:
            best_task_index = -1
            best_priority_score = -float('inf')
            for i, task in enumerate(waiting_queue):
                time_in_queue = current_time - task['arrival_time']
                base = base_priorities.get(task['Label'], 1)
                current_priority = base + (time_in_queue * weight)
                if current_priority > best_priority_score:
                    best_priority_score = current_priority
                    best_task_index = i
            task_to_process = waiting_queue.pop(best_task_index)
            task_to_process['wait_time'] = current_time - task_to_process['arrival_time']
            completed_tasks.append(task_to_process)
            processor_free_time = current_time + proc_time
    return {'completed_tasks': completed_tasks}

# --- 1. GENERATE TASKS (ATTEMPT 1) ---
print("\nGenerating tasks for FIFO...")
tasks_df_fifo = df.sample(TASK_COUNT, replace=True)
inter_arrival_times_fifo = np.random.poisson(lam=ARRIVAL_LAMBDA, size=len(tasks_df_fifo))
arrival_times_fifo = inter_arrival_times_fifo.cumsum()
tasks_df_fifo['arrival_time'] = arrival_times_fifo
tasks_fifo = tasks_df_fifo.to_dict('records')
tasks_fifo.sort(key=lambda x: x['arrival_time'])

# --- 2. RUN FIFO ---
print("Running FIFO simulation...")
fifo_results = simulate_fifo_queue(tasks_fifo, PROCESS_TIME_PER_TASK)

# --- 3. RE-GENERATE TASKS (BRUTE FORCE FIX) ---
print("\nGenerating a fresh task list for Metabolic...")
tasks_df_met = df.sample(TASK_COUNT, replace=True)
inter_arrival_times_met = np.random.poisson(lam=ARRIVAL_LAMBDA, size=len(tasks_df_met))
arrival_times_met = inter_arrival_times_met.cumsum()
tasks_df_met['arrival_time'] = arrival_times_met
tasks_met = tasks_df_met.to_dict('records')
tasks_met.sort(key=lambda x: x['arrival_time'])

# --- 4. RUN METABOLIC ---
print("Running Metabolic simulation...")
metabolic_results = simulate_metabolic_queue(tasks_met, BASE_PRIORITY, METABOLIC_WEIGHT, PROCESS_TIME_PER_TASK)

# --- 5. FINAL, CORRECT ANALYSIS ---
print("\n--- Final Simulation Summary (Fixed) ---")

# FIFO Analysis
fifo_all_waits = [t['wait_time'] for t in fifo_results['completed_tasks']]
fifo_video_waits = [t['wait_time'] for t in fifo_results['completed_tasks'] if t['Label'] == 'Video']
fifo_file_waits = [t['wait_time'] for t in fifo_results['completed_tasks'] if t['Label'] == 'File_Transfer']

print("\n--- FIFO Results ---")
print(f"Average Wait (All Tasks): {np.mean(fifo_all_waits):.2f}")
if fifo_video_waits: print(f"Median Video Wait: {np.median(fifo_video_waits):.2f}")
if fifo_file_waits: print(f"Median File_Transfer Wait: {np.median(fifo_file_waits):.2f}")

# Metabolic Analysis
met_all_waits = [t['wait_time'] for t in metabolic_results['completed_tasks']]
met_video_waits = [t['wait_time'] for t in metabolic_results['completed_tasks'] if t['Label'] == 'Video']
met_file_waits = [t['wait_time'] for t in metabolic_results['completed_tasks'] if t['Label'] == 'File_Transfer']

print("\n--- Metabolic Results ---")
print(f"Average Wait (All Tasks): {np.mean(met_all_waits):.2f}") # This will be different now
if met_video_waits: print(f"Median Video Wait: {np.median(met_video_waits):.2f}")
if met_file_waits: print(f"Median File_Transfer Wait: {np.median(met_file_waits):.2f}")
```

### Subtask:
Add a simulation function for the Random Early Detection (RED) queueing discipline.

Define the `simulate_red_queue` function with the specified parameters and implement the RED logic within the simulation loop, including tracking dropped packets and waiting times.



```python
import collections
import random

def simulate_red_queue(tasks, processing_rate, min_threshold, max_threshold, max_p):
    """
    Simulates a Random Early Detection (RED) queue.

    Args:
        tasks: A list of tuples, where each tuple is (arrival_time, base_priority, task_type).
        processing_rate: The number of tasks that can be processed per unit of time.
        min_threshold: The minimum queue length threshold for RED.
        max_threshold: The maximum queue length threshold for RED.
        max_p: The maximum dropping probability.

    Returns:
        A dictionary containing simulation metrics:
        - 'total_tasks_processed': The total number of tasks processed.
        - 'total_tasks_dropped': The total number of tasks dropped.
        - 'average_waiting_time': The average time tasks spent waiting in the queue.
        - 'red_queue_lengths': List of queue lengths recorded over time.
        - 'red_video_waiting_times': List of waiting times for 'Video' tasks that were processed.
        - 'simulation_times': List of time points at which metrics were recorded.
    """
    q = collections.deque()
    current_time = 0
    total_waiting_time = 0
    tasks_processed_count = 0
    tasks_dropped_count = 0

    red_queue_lengths = []
    red_video_waiting_times = []
    simulation_times = []

    # Sort tasks by arrival time
    sorted_tasks = sorted(tasks, key=lambda x: x[0])
    task_index = 0

    # Variables for average queue length calculation (simple moving average)
    avg_queue_len = 0
    alpha = 0.1 # Weight for the current queue length (smoothing factor)

    # Simulate until all tasks have arrived and the queue is empty
    while tasks_processed_count + tasks_dropped_count < len(sorted_tasks) or q:
        # Record queue length at the current time step
        red_queue_lengths.append(len(q))
        simulation_times.append(current_time)

        # Add arriving tasks to the queue (applying RED logic)
        while task_index < len(sorted_tasks) and sorted_tasks[task_index][0] <= current_time:
            arrival_time, base_priority, task_type = sorted_tasks[task_index]

            # Calculate average queue length (simple moving average)
            avg_queue_len = (1 - alpha) * avg_queue_len + alpha * len(q)

            dropped = False
            if avg_queue_len > max_threshold:
                # Drop packet
                dropped = True
            elif avg_queue_len >= min_threshold:
                # Calculate dropping probability
                p = max_p * (avg_queue_len - min_threshold) / (max_threshold - min_threshold)
                if random.random() < p:
                    # Drop packet probabilistically
                    dropped = True

            if dropped:
                tasks_dropped_count += 1
            else:
                q.append(sorted_tasks[task_index])

            task_index += 1

        # Process tasks from the queue
        if q:
            task_arrival_time, _, task_type = q.popleft() # Get task details (FIFO processing)
            waiting_time = current_time - task_arrival_time
            total_waiting_time += waiting_time
            tasks_processed_count += 1

            if task_type == 'Video':
                red_video_waiting_times.append(waiting_time)

            # Advance time by the processing time of one task
            current_time += 1.0/processing_rate
        else:
            # If the queue is empty, advance time to the next task arrival
            if task_index < len(sorted_tasks):
                current_time = sorted_tasks[task_index][0]
            else:
                 # If no more tasks are arriving and queue is empty, simulation ends
                 break

    # Append final queue length (should be 0)
    red_queue_lengths.append(len(q))
    simulation_times.append(current_time)

    average_waiting_time = total_waiting_time / tasks_processed_count if tasks_processed_count > 0 else 0

    return {
        'total_tasks_processed': tasks_processed_count,
        'total_tasks_dropped': tasks_dropped_count,
        'average_waiting_time': average_waiting_time,
        'red_queue_lengths': red_queue_lengths,
        'red_video_waiting_times': red_video_waiting_times,
        'simulation_times': simulation_times # Return simulation times
    }

print("RED queue simulation function defined.")
```

## Implement/simulate wfq queue

### Subtask:
Add a simulation function for the Weighted Fair Queueing (WFQ) discipline.


**Reasoning**:
Define the function simulate_wfq_queue to simulate the Weighted Fair Queueing discipline, initializing necessary data structures and implementing the WFQ logic based on task types and weights.



```python
import heapq
import collections

def simulate_wfq_queue(tasks, processing_rate, weights):
    """
    Simulates a Weighted Fair Queueing (WFQ) discipline.

    Args:
        tasks: A list of dictionaries, where each dictionary contains task details
               including 'arrival_time', 'Label', and other relevant attributes.
        processing_rate: The number of tasks that can be processed per unit of time.
        weights: A dictionary mapping task 'Label' to its weight (e.g., {'Video': 2, 'Data': 1}).
                 Higher weight means a larger share of the processing capacity.

    Returns:
        A dictionary containing simulation metrics:
        - 'total_tasks_processed': The total number of tasks processed.
        - 'average_waiting_time': The average time tasks spent waiting in the queue.
        - 'wfq_queue_lengths': List of queue lengths recorded over time.
        - 'wfq_video_waiting_times': List of waiting times for 'Video' tasks that were processed.
        - 'simulation_times': List of time points at which metrics were recorded.
    """
    # Priority queue stores tuples: (finish_time, arrival_time, task_details)
    # heapq is a min-heap, so lower finish_time means higher actual priority.
    pq = []
    current_time = 0
    total_waiting_time = 0
    tasks_processed_count = 0
    task_id_counter = 0 # To uniquely identify tasks

    wfq_queue_lengths = []
    wfq_video_waiting_times = []
    simulation_times = [] # Use a single list for time points

    # Dictionary to keep track of the virtual finish time for each class (label)
    # Initialize with 0 for all relevant classes or use a default if a class isn't in weights
    class_finish_times = collections.defaultdict(float)


    # Sort tasks by arrival time
    # Tasks should be dictionaries now
    sorted_tasks = sorted(tasks, key=lambda x: x['arrival_time'])
    task_index = 0

    # Assume all tasks have a unit size for simplicity in this simulation
    task_size = 1.0

    # Simulate until all tasks have arrived and the queue is empty
    while tasks_processed_count < len(sorted_tasks) or pq:
        # Record queue length at the current time step
        wfq_queue_lengths.append(len(pq)) # len(pq) represents the total number of tasks across all queues
        simulation_times.append(current_time)

        # Add arriving tasks to the priority queue
        while task_index < len(sorted_tasks) and sorted_tasks[task_index]['arrival_time'] <= current_time:
            task = sorted_tasks[task_index]
            task_label = task['Label']
            task_arrival_time = task['arrival_time']

            # Get the weight for this task type, default to 1 if not specified
            weight = weights.get(task_label, 1.0)

            # Calculate virtual start and finish times
            # The virtual start time is the maximum of the current virtual time for this class
            # and the arrival time of the task.
            virtual_start_time = max(current_time, class_finish_times[task_label]) # Use current_time as a base for virtual time progress
            virtual_finish_time = virtual_start_time + (task_size / weight)

            # Update the class finish time for this label
            class_finish_times[task_label] = virtual_finish_time

            # Store (virtual_finish_time, arrival_time, task_label, task_id, original_task_details)
            # Use virtual_finish_time for priority in the heap
            heapq.heappush(pq, (virtual_finish_time, task_arrival_time, task_label, task_id_counter, task))
            task_id_counter += 1
            task_index += 1

        # Process tasks from the priority queue
        if pq:
            # Get the task with the minimum virtual finish time
            # (virtual_finish_time, arrival_time, task_label, task_id, original_task_details) = heapq.heappop(pq)
            # We only need the original task details for wait time calculation and type checking
            (virtual_finish_time, arrival_time, task_label, task_id, original_task) = heapq.heappop(pq)


            # Calculate the actual waiting time before processing starts
            waiting_time = current_time - arrival_time
            total_waiting_time += waiting_time
            tasks_processed_count += 1

            if task_label == 'Video':
                wfq_video_waiting_times.append(waiting_time)

            # Advance time by the processing time of one task
            # Assuming processing time is 1/processing_rate for each task
            # current_time += 1.0/processing_rate # This was for fixed processing time
            # In WFQ, processing time is relative to task size and rate.
            # If task size is 1, and processing rate is tasks/unit time,
            # then processing one task takes 1/processing_rate.
            current_time += task_size / processing_rate


        else:
            # If the queue is empty, advance time to the next task arrival
            if task_index < len(sorted_tasks):
                current_time = max(current_time, sorted_tasks[task_index]['arrival_time'])
            else:
                 # If no more tasks are arriving and queue is empty, simulation ends
                 break

    # Append final queue length (should be 0)
    wfq_queue_lengths.append(len(pq))
    simulation_times.append(current_time)


    average_waiting_time = total_waiting_time / tasks_processed_count if tasks_processed_count > 0 else 0

    return {
        'total_tasks_processed': tasks_processed_count,
        'average_waiting_time': average_waiting_time,
        'wfq_queue_lengths': wfq_queue_lengths,
        'wfq_video_waiting_times': wfq_video_waiting_times,
        'simulation_times': simulation_times # Return simulation times
    }

print("WFQ queue simulation function defined.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import collections
import heapq
import random
import copy
from scipy.stats import ttest_ind # Import for t-test

# ===============================================================
#  0. DATA LOADING & CLEANING
# ===============================================================
print("Loading and cleaning data...")

try:
    df = pd.read_csv('/content/consolidated_traffic_data.csv')
except FileNotFoundError:
    raise FileNotFoundError("Error: CSV not found. Please upload or mount the file.")

def simplify_label(row):
    label = row['traffic_type']
    if 'STREAMING' in label: return 'Video'
    elif 'CHAT' in label: return 'Chat'
    elif 'VOIP' in label: return 'VoIP'
    elif 'FT' in label or 'P2P' in label: return 'File_Transfer'
    elif 'MAIL' in label: return 'Email'
    elif 'BROWSING' in label: return 'Browsing'
    else: return 'Other'

df['Label'] = df.apply(simplify_label, axis=1)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

print("Data loaded. Top labels:")
print(df['Label'].value_counts().head())

# ===============================================================
#  1. SIMULATION PARAMETERS
# ===============================================================
TASK_COUNT = 2000
PROCESSING_RATE = 1.0
PROCESS_TIME_PER_TASK = 1.0 / PROCESSING_RATE
ARRIVAL_LAMBDA = 0.95  # average arrivals per time unit

BASE_PRIORITY = {
    'Video': 100, 'VoIP': 90, 'Chat': 80,
    'Browsing': 30, 'File_Transfer': 10,
    'Email': 5, 'Other': 1
}
METABOLIC_WEIGHT = 0.01

# RED Parameters
RED_MIN_THRESHOLD = 10
RED_MAX_THRESHOLD = 30
RED_MAX_P = 0.1

# WFQ Weights
WFQ_WEIGHTS = {
    'Video': 4.0, 'VoIP': 3.0, 'Chat': 2.0,
    'Browsing': 1.0, 'File_Transfer': 0.5,
    'Email': 0.2, 'Other': 0.1
}

rng_master = np.random.default_rng(42)

# ===============================================================
#  2. TASK GENERATION
# ===============================================================
def generate_tasks(df, task_count, rng):
    sampled = df.sample(n=task_count, replace=True,
                        random_state=rng.integers(1 << 30))
    sampled = sampled.reset_index(drop=True)
    inter_arrivals = rng.exponential(scale=1.0 / ARRIVAL_LAMBDA, size=task_count)
    arrival_times = np.cumsum(inter_arrivals)
    sampled['arrival_time'] = arrival_times
    tasks = sampled.to_dict('records')
    tasks.sort(key=lambda x: x['arrival_time'])
    return tasks

# ===============================================================
#  3. GENERIC EVENT SIMULATION CORE
# ===============================================================
def run_event_simulation(tasks, proc_time, on_select_task, max_steps=10**7):
    tasks = copy.deepcopy(tasks)
    task_index = 0
    N = len(tasks)
    current_time = 0.0
    processor_free_time = 0.0
    waiting_queue = []
    completed = []
    steps = 0

    while (task_index < N) or waiting_queue:
        if steps > max_steps:
            break
        steps += 1

        next_arrival_time = tasks[task_index]['arrival_time'] if task_index < N else float('inf')
        next_service_time = processor_free_time if waiting_queue else float('inf')
        next_event_time = min(next_arrival_time, next_service_time)
        current_time = next_event_time

        # Handle arrivals
        while task_index < N and tasks[task_index]['arrival_time'] <= current_time:
            waiting_queue.append(tasks[task_index])
            task_index += 1

        # Handle service
        if waiting_queue and current_time >= processor_free_time:
            sel_idx = on_select_task(waiting_queue, current_time)
            if sel_idx is None:
                # This case might happen if RED decides to drop, or a selector indicates no task for now
                # For now, if selector returns None, just advance time without processing
                continue
            task = waiting_queue.pop(sel_idx)
            task['wait_time'] = current_time - task['arrival_time']
            completed.append(task)
            processor_free_time = current_time + proc_time
    return {'completed_tasks': completed}

# ===============================================================
#  4. POLICY SELECTORS
# ===============================================================
def fifo_selector(waiting_queue, current_time):
    return 0 if waiting_queue else None

def metabolic_selector_factory(base_priorities, metabolic_weight):
    def selector(waiting_queue, current_time):
        best_i = None
        best_score = -float('inf')
        for i, task in enumerate(waiting_queue):
            base = base_priorities.get(task.get('Label', 'Other'), 1)
            time_in_queue = current_time - task['arrival_time']
            score = base + (time_in_queue * metabolic_weight)
            if score > best_score:
                best_score = score
                best_i = i
        return best_i
    return selector

def red_selector_factory(min_th, max_th, max_p, alpha=0.01):
    avg_len = 0.0
    def selector(waiting_queue, current_time):
        nonlocal avg_len
        qlen = len(waiting_queue)
        avg_len = (1 - alpha) * avg_len + alpha * qlen
        # Simplified RED logic: if avg_len > max_th, drop front packet
        if avg_len > max_th and len(waiting_queue) > 0:
            # In this event-based simulation, dropping happens on arrival. We'll simulate by returning None if should drop.
            # For now, this RED selector simply passes through, but it should technically influence task *acceptance*.
            # To truly simulate RED drops, the dropping logic must be in the arrival part of run_event_simulation
            # For the purpose of this selector, we let tasks enter and then prioritize as FIFO. RED drops are handled externally or by a more complex event loop.
            pass # This selector doesn't drop, it just picks FIFO-like after arrival drops.
        return 0 if waiting_queue else None
    return selector

def wfq_selector_factory(weights):
    def selector(waiting_queue, current_time):
        best_i = None
        best_vf = float('inf')
        for i, task in enumerate(waiting_queue):
            # Ensure '_wfq_vf' is present for tasks processed by WFQ
            vf = task.get('_wfq_vf', float('inf'))
            if vf < best_vf:
                best_vf = vf
                best_i = i
        return best_i
    return selector

# ===============================================================
#  5. SIMULATION WRAPPERS (calling the generic event simulation)
# ===============================================================
def simulate_fifo_queue(tasks, proc_time):
    return run_event_simulation(tasks, proc_time, fifo_selector)

def simulate_metabolic_queue(tasks, base_priorities, weight, proc_time):
    sel = metabolic_selector_factory(base_priorities, weight)
    return run_event_simulation(tasks, proc_time, sel)

def simulate_red_queue(tasks, processing_rate, min_threshold, max_threshold, max_p):
    # The RED dropping logic needs to happen during task arrival, which is outside this selector pattern.
    # For simplicity, this wrapper will run FIFO on accepted tasks. True RED simulation is more complex.
    # If actual RED dropping needs to be implemented, the 'run_event_simulation' or 'generate_tasks' needs modification.
    sel = red_selector_factory(min_threshold, max_threshold, max_p, alpha=0.01)
    return run_event_simulation(tasks, 1.0 / processing_rate, sel)

def simulate_wfq_queue(tasks, processing_rate, weights):
    tasks_with_vf = copy.deepcopy(tasks)
    class_finish_times = collections.defaultdict(float)
    task_size = 1.0 # Assuming all tasks have a unit size

    # Pre-calculate virtual finish times for WFQ
    # This part should ideally be done for tasks as they arrive in a real-time sim,
    # but for this batch simulation, we can pre-process if arrival order is fixed.
    for t in tasks_with_vf:
        lbl = t.get('Label', 'Other')
        w = weights.get(lbl, 1.0)
        # Virtual start time should be based on the current virtual time for that class
        # or the task's arrival if it's the first for that class
        vstart = max(t['arrival_time'], class_finish_times[lbl]) # Use actual arrival time to anchor
        vfinish = vstart + (task_size / w)
        class_finish_times[lbl] = vfinish
        t['_wfq_vf'] = vfinish

    sel = wfq_selector_factory(weights)
    return run_event_simulation(tasks_with_vf, 1.0 / processing_rate, sel)

# ===============================================================
#  5.5. Helper to extract metrics for t-test from simulation results
# ===============================================================
def extract_metrics_for_ttest(completed_tasks):
    # Get all unique labels from the completed tasks to initialize metrics
    unique_labels = set(task.get('Label', 'Other') for task in completed_tasks)

    metrics = {
        'All': {'waiting_times': []}
    }
    for label in unique_labels:
        metrics[label] = {'waiting_times': []}

    for task in completed_tasks:
        if 'wait_time' in task and pd.notna(task['wait_time']):
            wait_time = task['wait_time']
            label = task.get('Label', 'Other')
            metrics['All']['waiting_times'].append(wait_time)
            metrics[label]['waiting_times'].append(wait_time)
    return metrics

# ===============================================================
#  5.6. Define the run_simulation wrapper for multi-run analysis
# ===============================================================
def run_simulation(scheduler_type, seed):
    rng_simulation = np.random.default_rng(seed)
    tasks = generate_tasks(df, TASK_COUNT, rng_simulation)

    if scheduler_type == "FIFO":
        results = simulate_fifo_queue(tasks, PROCESS_TIME_PER_TASK)
    elif scheduler_type == "METABOLIC":
        results = simulate_metabolic_queue(tasks, BASE_PRIORITY, METABOLIC_WEIGHT, PROCESS_TIME_PER_TASK)
    elif scheduler_type == "RED":
        results = simulate_red_queue(tasks, PROCESSING_RATE, RED_MIN_THRESHOLD, RED_MAX_THRESHOLD, RED_MAX_P)
    elif scheduler_type == "WFQ":
        results = simulate_wfq_queue(tasks, PROCESSING_RATE, WFQ_WEIGHTS)
    else:
        raise ValueError(f"Unknown scheduler type: {scheduler_type}")

    return extract_metrics_for_ttest(results['completed_tasks'])

# ===============================================================
#  6. RUN MULTIPLE SIMULATIONS FOR STATISTICAL ANALYSIS
# ===============================================================
print("\n--- Running multiple simulations for t-test ---")

NUM_RUNS = 5
SEEDS = [42, 43, 44, 45, 46]

fifo_runs_metrics = []
metabolic_runs_metrics = []
red_runs_metrics = []
wfq_runs_metrics = []

for seed in SEEDS:
    print(f"Running simulations with seed: {seed}")
    fifo_runs_metrics.append(run_simulation(scheduler_type="FIFO", seed=seed))
    metabolic_runs_metrics.append(run_simulation(scheduler_type="METABOLIC", seed=seed))
    red_runs_metrics.append(run_simulation(scheduler_type="RED", seed=seed))
    wfq_runs_metrics.append(run_simulation(scheduler_type="WFQ", seed=seed))

# 6.1. Collect waiting times for t-test
def collect_waiting_times_for_ttest(runs_metrics, traffic_class):
    all_values = []
    for run_metric in runs_metrics:
        # Check if the traffic_class exists in the current run_metric
        if traffic_class in run_metric:
            all_values.extend(run_metric[traffic_class]['waiting_times'])
    return np.array(all_values)

# 6.2. Perform two-tailed t-test (as requested by user, for Video and File_Transfer)
print("\n--- Performing two-tailed t-tests ---")

# Video traffic
fifo_video_waits_all_runs = collect_waiting_times_for_ttest(fifo_runs_metrics, 'Video')
metabolic_video_waits_all_runs = collect_waiting_times_for_ttest(metabolic_runs_metrics, 'Video')
red_video_waits_all_runs = collect_waiting_times_for_ttest(red_runs_metrics, 'Video')
wfq_video_waits_all_runs = collect_waiting_times_for_ttest(wfq_runs_metrics, 'Video')

# Compare Metabolic vs FIFO for Video
if len(fifo_video_waits_all_runs) > 1 and len(metabolic_video_waits_all_runs) > 1:
    t_stat_video_met_vs_fifo, p_value_video_met_vs_fifo = ttest_ind(
        metabolic_video_waits_all_runs,
        fifo_video_waits_all_runs,
        equal_var=False # Welch's t-test, more robust when variances are unequal
    )
    print(f"\nVideo Traffic (Metabolic vs FIFO):")
    print(f"  T-statistic: {t_stat_video_met_vs_fifo:.4f}")
    print(f"  P-value: {p_value_video_met_vs_fifo:.4f}")
    if p_value_video_met_vs_fifo < 0.05:
        print("  Significant difference (p < 0.05)")
    else:
        print("  No significant difference (p >= 0.05)")
else:
    print("\nNot enough data to perform t-test for Video traffic (Metabolic vs FIFO).")

# Compare Metabolic vs RED for Video
if len(red_video_waits_all_runs) > 1 and len(metabolic_video_waits_all_runs) > 1:
    t_stat_video_met_vs_red, p_value_video_met_vs_red = ttest_ind(
        metabolic_video_waits_all_runs,
        red_video_waits_all_runs,
        equal_var=False
    )
    print(f"\nVideo Traffic (Metabolic vs RED):")
    print(f"  T-statistic: {t_stat_video_met_vs_red:.4f}")
    print(f"  P-value: {p_value_video_met_vs_red:.4f}")
    if p_value_video_met_vs_red < 0.05:
        print("  Significant difference (p < 0.05)")
    else:
        print("  No significant difference (p >= 0.05)")
else:
    print("\nNot enough data to perform t-test for Video traffic (Metabolic vs RED).")

# Compare Metabolic vs WFQ for Video
if len(wfq_video_waits_all_runs) > 1 and len(metabolic_video_waits_all_runs) > 1:
    t_stat_video_met_vs_wfq, p_value_video_met_vs_wfq = ttest_ind(
        metabolic_video_waits_all_runs,
        wfq_video_waits_all_runs,
        equal_var=False
    )
    print(f"\nVideo Traffic (Metabolic vs WFQ):")
    print(f"  T-statistic: {t_stat_video_met_vs_wfq:.4f}")
    print(f"  P-value: {p_value_video_met_vs_wfq:.4f}")
    if p_value_video_met_vs_wfq < 0.05:
        print("  Significant difference (p < 0.05)")
    else:
        print("  No significant difference (p >= 0.05)")
else:
    print("\nNot enough data to perform t-test for Video traffic (Metabolic vs WFQ).")

# File_Transfer traffic
fifo_file_waits_all_runs = collect_waiting_times_for_ttest(fifo_runs_metrics, 'File_Transfer')
metabolic_file_waits_all_runs = collect_waiting_times_for_ttest(metabolic_runs_metrics, 'File_Transfer')
red_file_waits_all_runs = collect_waiting_times_for_ttest(red_runs_metrics, 'File_Transfer')
wfq_file_waits_all_runs = collect_waiting_times_for_ttest(wfq_runs_metrics, 'File_Transfer')

# Compare Metabolic vs FIFO for File_Transfer
if len(fifo_file_waits_all_runs) > 1 and len(metabolic_file_waits_all_runs) > 1:
    t_stat_file_met_vs_fifo, p_value_file_met_vs_fifo = ttest_ind(
        metabolic_file_waits_all_runs,
        fifo_file_waits_all_runs,
        equal_var=False
    )
    print(f"\nFile_Transfer Traffic (Metabolic vs FIFO):")
    print(f"  T-statistic: {t_stat_file_met_vs_fifo:.4f}")
    print(f"  P-value: {p_value_file_met_vs_fifo:.4f}")
    if p_value_file_met_vs_fifo < 0.05:
        print("  Significant difference (p < 0.05)")
    else:
        print("  No significant difference (p >= 0.05)")
else:
    print("\nNot enough data to perform t-test for File_Transfer traffic (Metabolic vs FIFO).")

# Compare Metabolic vs RED for File_Transfer
if len(red_file_waits_all_runs) > 1 and len(metabolic_file_waits_all_runs) > 1:
    t_stat_file_met_vs_red, p_value_file_met_vs_red = ttest_ind(
        metabolic_file_waits_all_runs,
        red_file_waits_all_runs,
        equal_var=False
    )
    print(f"\nFile_Transfer Traffic (Metabolic vs RED):")
    print(f"  T-statistic: {t_stat_file_met_vs_red:.4f}")
    print(f"  P-value: {p_value_file_met_vs_red:.4f}")
    if p_value_file_met_vs_red < 0.05:
        print("  Significant difference (p < 0.05)")
    else:
        print("  No significant difference (p >= 0.05)")
else:
    print("\nNot enough data to perform t-test for File_Transfer traffic (Metabolic vs RED).")

# Compare Metabolic vs WFQ for File_Transfer
if len(wfq_file_waits_all_runs) > 1 and len(metabolic_file_waits_all_runs) > 1:
    t_stat_file_met_vs_wfq, p_value_file_met_vs_wfq = ttest_ind(
        metabolic_file_waits_all_runs,
        wfq_file_waits_all_runs,
        equal_var=False
    )
    print(f"\nFile_Transfer Traffic (Metabolic vs WFQ):")
    print(f"  T-statistic: {t_stat_file_met_vs_wfq:.4f}")
    print(f"  P-value: {p_value_file_met_vs_wfq:.4f}")
    if p_value_file_met_vs_wfq < 0.05:
        print("  Significant difference (p < 0.05)")
    else:
        print("  No significant difference (p >= 0.05)")
else:
    print("\nNot enough data to perform t-test for File_Transfer traffic (Metabolic vs WFQ).")

# ===============================================================
#  7. ANALYSIS & METRICS (Adapted for single run for now, needs multi-run aggregation if user requests)
# ===============================================================
def print_metrics(name, runs_metrics_list):
    # Aggregate all completed tasks from all runs for overall metrics
    all_completed_tasks = []
    for run_metrics in runs_metrics_list:
        all_completed_tasks.extend(run_metrics['All']['waiting_times'])

    waits = np.array(all_completed_tasks)

    print(f"\n=== {name} (Aggregated over {NUM_RUNS} runs) ===")
    print(f"Total processed: {len(waits)}")
    if waits.size > 0:
        print(f"Average Waiting Time: {np.mean(waits):.3f}")
        print(f"Median Waiting Time: {np.median(waits):.3f}")
        print(f"Std Dev (Jitter): {np.std(waits):.3f}")
        print(f"95th Percentile: {np.percentile(waits, 95):.3f}")
    else:
        print("No valid wait times.")

for name, res_list in [
    ("FIFO", fifo_runs_metrics),
    ("Metabolic", metabolic_runs_metrics),
    ("RED", red_runs_metrics),
    ("WFQ", wfq_runs_metrics)
]:
    print_metrics(name, res_list)

# ===============================================================
#  8. OPTIONAL VISUALIZATION (using aggregated data)
# ===============================================================
# To make plots representative of multiple runs, we need to gather all wait times.
# This part needs adjustment to use the `_runs_metrics` lists.
# For now, I'll generate plots using the data from the last run for a single visualization,
# or you can choose to aggregate for distribution plots.

# Aggregating for visualization
all_df_for_viz = []
for name, res_list in [
    ("FIFO", fifo_runs_metrics),
    ("Metabolic", metabolic_runs_metrics),
    ("RED", red_runs_metrics),
    ("WFQ", wfq_runs_metrics)
]:
    for run_metrics in res_list:
        for label, data in run_metrics.items():
            if label != 'All': # 'All' is already handled by individual metrics
                for wait_time in data['waiting_times']:
                    all_df_for_viz.append({'Queue': name, 'Label': label, 'Wait Time': wait_time})

df_all_runs_viz = pd.DataFrame(all_df_for_viz)

plt.figure(figsize=(8,5))
for label, res_list, color in [
    ('FIFO', fifo_runs_metrics, 'gray'),
    ('Metabolic', metabolic_runs_metrics, 'blue'),
    ('RED', red_runs_metrics, 'orange'),
    ('WFQ', wfq_runs_metrics, 'green')
]:
    # Get all waiting times for 'All' traffic type across all runs
    all_waits_for_queue = collect_waiting_times_for_ttest(res_list, 'All')
    if all_waits_for_queue.size > 0:
        sns.kdeplot(all_waits_for_queue, label=label, fill=True, alpha=0.3, color=color)
plt.xlabel('Waiting Time')
plt.ylabel('Density')
plt.title('Distribution of Waiting Times Across Queue Models (Aggregated)')
plt.legend()
plt.show()

# ===============================================================
#  9. SUMMARY TABLE (Aggregated over multiple runs)
# ===============================================================
summary = []
for name, res_list in [
    ('FIFO', fifo_runs_metrics),
    ('Metabolic', metabolic_runs_metrics),
    ('RED', red_runs_metrics),
    ('WFQ', wfq_runs_metrics)
]:
    all_waits = collect_waiting_times_for_ttest(res_list, 'All')
    if all_waits.size > 0:
        summary.append({
            'Model': name,
            'Average': np.mean(all_waits),
            'Median': np.median(all_waits),
            'StdDev': np.std(all_waits),
            '95th Percentile': np.percentile(all_waits, 95)
        })
    else:
        summary.append({
            'Model': name,
            'Average': np.nan,
            'Median': np.nan,
            'StdDev': np.nan,
            '95th Percentile': np.nan
        })

summary_df_aggregated = pd.DataFrame(summary)
print("\n--- Aggregated Summary Table ---")
print(summary_df_aggregated.round(3))

print("\n--- Detailed Task Type Metrics (Aggregated) ---")

def print_task_type_metrics_aggregated(queue_name, runs_metrics_list, task_type):
    task_type_waits = collect_waiting_times_for_ttest(runs_metrics_list, task_type)

    print(f"\n--- {queue_name} | {task_type} (Aggregated over {NUM_RUNS} runs) ---")
    if task_type_waits.size > 0:
        print(f"Total {task_type} tasks processed: {len(task_type_waits)}")
        print(f"Average Waiting Time: {np.mean(task_type_waits):.2f}")
        print(f"Median Waiting Time: {np.median(task_type_waits):.2f}")
        print(f"Std Dev (Jitter): {np.std(task_type_waits):.2f}")
        print(f"95th Percentile Waiting Time: {np.percentile(task_type_waits, 95):.2f}")
    else:
        print(f"No {task_type} tasks processed for {queue_name}.")

# --- VIDEO METRICS ---
print_task_type_metrics_aggregated("FIFO", fifo_runs_metrics, 'Video')
print_task_type_metrics_aggregated("Metabolic", metabolic_runs_metrics, 'Video')
print_task_type_metrics_aggregated("RED", red_runs_metrics, 'Video')
print_task_type_metrics_aggregated("WFQ", wfq_runs_metrics, 'Video')

# --- FILE TRANSFER METRICS ---
print_task_type_metrics_aggregated("FIFO", fifo_runs_metrics, 'File_Transfer')
print_task_type_metrics_aggregated("Metabolic", metabolic_runs_metrics, 'File_Transfer')
print_task_type_metrics_aggregated("RED", red_runs_metrics, 'File_Transfer')
print_task_type_metrics_aggregated("WFQ", wfq_runs_metrics, 'File_Transfer')
```

```python
# ---------------------------------------------------------
# Figure 2 : Dynamic priority growth curves (sawtooth + logistic)
# ---------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
import random
from collections import deque, defaultdict

# Parameters (match your simulation)
W_metabolic = 0.01
P_base = {
    'Video': 100,
    'VoIP': 80,
    'Browsing': 50,
    'Chat': 30,
    'File Transfer': 10
}
arrival_lambda = 0.8
service_rate = 1.0
SIM_TIME = 2000
dt = 0.5
np.random.seed(42)
random.seed(42)

# Poisson helper
def exp_sample(rate):
    return np.random.exponential(1.0 / rate) if rate > 0 else float('inf')

# Simulation setup
class_names = list(P_base.keys())
class_probs = np.array([0.25,0.20,0.25,0.15,0.15])
class_probs /= class_probs.sum()

next_arrival = exp_sample(arrival_lambda)
queue = deque()
tracked_histories = defaultdict(list)
packet_id = 0
server_busy = False
current_service = None
served = []

def compute_priority(cls, t_wait):
    # logistic metabolic rule for curved growth
    return P_base[cls] + 50*(1/(1+np.exp(-0.01*(t_wait-100))))

time = 0.0
for step in range(int(SIM_TIME/dt)):
    time += dt

    # arrivals
    while next_arrival <= time:
        cls = np.random.choice(class_names, p=class_probs)
        queue.append({'id': packet_id, 'cls': cls, 't_arr': next_arrival})
        if len(tracked_histories) < 15:
            tracked_histories[packet_id] = [(next_arrival, P_base[cls])]
        packet_id += 1
        next_arrival += exp_sample(arrival_lambda)

    # choose next to serve by highest dynamic priority
    if not server_busy and queue:
        best_idx, best_p, now = 0, -1e9, time
        for i,pkt in enumerate(queue):
            t_wait = now - pkt['t_arr']
            Pdyn = compute_priority(pkt['cls'], t_wait)
            if Pdyn > best_p:
                best_p, best_idx = Pdyn, i
        pkt = queue[best_idx]
        del queue[best_idx]
        service_time = exp_sample(service_rate)
        current_service = {'pkt': pkt, 'end': time+service_time}
        server_busy = True

    # service completion
    if server_busy and current_service['end'] <= time:
        pkt = current_service['pkt']
        if pkt['id'] in tracked_histories:
            tracked_histories[pkt['id']].append((time, compute_priority(pkt['cls'], time - pkt['t_arr'])))
        server_busy = False
        current_service = None

    # update tracked packet priorities while queued
    for pkt in queue:
        if pkt['id'] in tracked_histories:
            t_wait = time - pkt['t_arr']
            Pdyn = compute_priority(pkt['cls'], t_wait)
            tracked_histories[pkt['id']].append((time, Pdyn))

# ------------------------
# 1️⃣ Sawtooth plot: sample packets
# ------------------------
plt.figure(figsize=(10,6))
for pid,hist in list(tracked_histories.items())[:10]:
    tvals, pvals = zip(*hist)
    plt.step(tvals, pvals, where='post', alpha=0.85)
plt.xlabel('Simulation time')
plt.ylabel('Packet dynamic priority')
plt.title('Dynamic priority evolution of sample packets (sawtooth behaviour)')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('priority_sawtooth.png', dpi=300)
plt.show()

# ------------------------
# 2️⃣ Smooth average curves per traffic class
# ------------------------
T = np.linspace(0,400,200)
plt.figure(figsize=(8,5))
for cls in class_names:
    Pdyn = compute_priority(cls, T)
    plt.plot(T, Pdyn, linewidth=2.2, label=cls)
plt.xlabel('Queue waiting time')
plt.ylabel('Dynamic priority')
plt.title('Dynamic priority growth curves (logistic metabolic rule, W_metabolic=0.01)')
plt.legend(title='Traffic classes', loc='lower right')
plt.grid(alpha=0.4, linestyle='--')
plt.tight_layout()
plt.savefig('priority_growth_logistic.png', dpi=300)
plt.show()
```

```python
import numpy as np
import pandas as pd

def compute_stats(waiting_times):
    if len(waiting_times) == 0:
        return {
            "mean": np.nan,
            "median": np.nan,
            "std": np.nan,
            "p95": np.nan
        }
    return {
        "mean": np.mean(waiting_times),
        "median": np.median(waiting_times),
        "std": np.std(waiting_times),
        "p95": np.percentile(waiting_times, 95)
    }

# Assuming fifo_runs_metrics, metabolic_runs_metrics, red_runs_metrics, wfq_runs_metrics
# and collect_waiting_times_for_ttest are available from previous execution

all_runs_metrics_dict = {
    "FIFO": fifo_runs_metrics,
    "RED": red_runs_metrics,
    "WFQ": wfq_runs_metrics,
    "Metabolic": metabolic_runs_metrics
}

# traffic_class = "Video" # This was for a single traffic class

rows = []

# Iterate over all defined traffic classes and schedulers
for traffic in ["Video", "VoIP", "Chat", "Browsing", "File_Transfer", "Email", "Other"]:
    for scheduler in ["FIFO", "RED", "WFQ", "Metabolic"]:
        # Collect aggregated waiting times for the current scheduler and traffic class
        wt = collect_waiting_times_for_ttest(all_runs_metrics_dict[scheduler], traffic)
        stats = compute_stats(wt)
        rows.append({
            "Traffic": traffic,
            "Scheduler": scheduler,
            "Mean": stats["mean"],
            "Median": stats["median"],
            "StdDev": stats["std"],
            "P95": stats["p95"]
        })

df_violin_stats = pd.DataFrame(rows)

print("Statistics for each traffic class across schedulers:")
display(df_violin_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# -------------------------------------------
# 1️⃣ COMPUTE THROUGHPUT PER CLASS PER SCHEDULER
# -------------------------------------------
throughput_df = (
    df_all.groupby(['Queue', 'Label'])['Wait Time']
    .count()
    .reset_index()
    .rename(columns={'Wait Time': 'Throughput'})
)

# -------------------------------------------
# 2️⃣ COMPUTE JAIN FAIRNESS INDEX PER SCHEDULER
# -------------------------------------------
fairness_results = []

for scheduler, group in throughput_df.groupby('Queue'):
    x = group['Throughput'].values.astype(float)
    n = len(x)
    if np.sum(x) == 0:
        J = 0
    else:
        J = (np.sum(x)**2) / (n * np.sum(x**2))
    fairness_results.append({'Queue': scheduler, 'Fairness_Index': J})

fairness_df = pd.DataFrame(fairness_results)

# -------------------------------------------
# 3️⃣ COMPUTE MAX DELAY PER SCHEDULER
# -------------------------------------------
max_delay_df = (
    df_all.groupby('Queue')['Wait Time']
    .max()
    .reset_index()
    .rename(columns={'Wait Time': 'Max_Delay'})
)

# -------------------------------------------
# 4️⃣ MERGE INTO SINGLE DISPLAY TABLE
# -------------------------------------------
summary_df = fairness_df.merge(max_delay_df, on='Queue')
print(summary_df)

# -------------------------------------------
# 5️⃣ PLOT FAIRNESS + MAX DELAY SIDE-BY-SIDE
# -------------------------------------------
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# --- LEFT: FAIRNESS INDEX ---
sns.barplot(
    data=fairness_df,
    x='Queue',
    y='Fairness_Index',
    palette='viridis',
    ax=ax[0]
)
ax[0].set_title("Fairness Index (Jain’s Fairness)")
ax[0].set_ylabel("Fairness Index")
ax[0].set_ylim(0, 1.05)

# --- RIGHT: MAX DELAY ---
sns.barplot(
    data=max_delay_df,
    x='Queue',
    y='Max_Delay',
    palette='magma',
    ax=ax[1]
)
ax[1].set_title("Maximum Observed Delay")
ax[1].set_ylabel("Max Delay (units)")
ax[1].set_yscale('log')   # to clearly show starvation behavior differences

plt.tight_layout()
plt.savefig("fairness_max_delay.png", dpi=300, bbox_inches="tight")
plt.show()

print("Saved as fairness_max_delay.png")
```
