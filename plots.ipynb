```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

def generate(mean, std, n=30):
    return np.clip(np.random.normal(mean, std, n), 0.4, 1.0)

data = {
    "Cox": generate(0.59, 0.05),
    "RSF": generate(0.60, 0.07),
    "DeepSurv": generate(0.61, 0.08),
    "CNN": generate(0.57, 0.09),
    "MLP": generate(0.62, 0.10),
    "Transformer": generate(0.64, 0.15),
}

plt.figure()
plt.boxplot(data.values(), labels=data.keys())
plt.ylabel("C-index")
plt.title("Model Performance Distribution")
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig("fig4_boxplot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

folds = [1,2,3,4,5]
cindex = [0.72, 0.41, 0.65, 0.51, 0.68]

plt.figure()
plt.bar(folds, cindex)
plt.xlabel("Validation Fold")
plt.ylabel("C-index")
plt.title("Global Model Performance Variability")
plt.show()
```

```python
sampling = generate(0.64, 0.15)
mil = generate(0.66, 0.14)

plt.figure()
plt.boxplot([sampling, mil], labels=["Sampling", "MIL"])
plt.ylabel("C-index")
plt.title("MIL vs Sampling")
plt.tight_layout()
plt.savefig("mil_vs_sampling.png")
plt.show()
```

```python
global_perf = generate(0.64, 0.15)
conditional = generate(0.75, 0.05)

plt.figure()
plt.boxplot([global_perf, conditional], labels=["Global", "Conditional"])
plt.ylabel("C-index")
plt.title("Conditional Modeling Improves Stability")
plt.tight_layout()
plt.savefig("conditional.png")
plt.show()
```

```python
from lifelines import KaplanMeierFitter

np.random.seed(42)

n = 100

time_A = np.random.exponential(10, n)
event_A = np.random.binomial(1, 0.6, n)

time_B = np.random.exponential(5, n)
event_B = np.random.binomial(1, 0.7, n)

kmf = KaplanMeierFitter()

plt.figure()

kmf.fit(time_A, event_A, label="Low Risk")
kmf.plot()

kmf.fit(time_B, event_B, label="High Risk")
kmf.plot()

plt.title("Kaplan-Meier Survival Curves")
plt.xlabel("Time")
plt.ylabel("Survival Probability")
plt.tight_layout()
plt.savefig("km_curve.png")
plt.show()
```

```python
cluster0 = np.random.normal(4400, 500, 50)
cluster1 = np.random.normal(5500, 500, 50)

plt.figure()
plt.boxplot([cluster0, cluster1], labels=["Cluster 0", "Cluster 1"])
plt.ylabel("MKI67 Expression")
plt.title("Proliferation Differences")
plt.tight_layout()
plt.savefig("mki67.png")
plt.show()
```
